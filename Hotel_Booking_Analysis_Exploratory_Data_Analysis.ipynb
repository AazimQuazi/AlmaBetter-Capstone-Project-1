# **Questions**
1) What is the overall distribution of bookings?
---



```python
# Convert date columns to datetime format
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])
#df['arrival_date'] = pd.to_datetime(df['arrival_date'])
df['arrival_date'] = pd.to_datetime(df[['arrival_date_year', 'arrival_date_month', 'arrival_date_day_of_month']].astype(str).agg('-'.join, axis=1))



 #Extract month and day of the week for better analysis

df['arrival_month'] = df['arrival_date'].dt.month
df['day_of_week'] = df['arrival_date'].dt.day_name()
```

```python
# Calculate the total number of bookings per month
monthly_bookings = df.groupby(df['arrival_date'].dt.to_period("M")).size()

# Calculate the total number of bookings per day of the week
daily_bookings = df.groupby(df['day_of_week']).size()
```

```python
# Plot monthly distribution
plt.figure(figsize=(10, 6))
sns.lineplot(data=monthly_bookings)
plt.title('Monthly Distribution of Bookings')
plt.xlabel('Month')
plt.ylabel('Number of Bookings')
plt.show()


```

```python
# Plot daily distribution
plt.figure(figsize=(10, 6))
sns.barplot(x=daily_bookings.index, y=daily_bookings)
plt.title('Daily Distribution of Bookings')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Bookings')
plt.show()
```

# **2) What is the cancellation rate?**

```python
# The column 'is_canceled' contains binary values (0 for not canceled, 1 for canceled)
cancellation_rate = df['is_canceled'].mean() * 100
print(f"Cancellation Rate: {cancellation_rate:.2f}%")   #two floating point number

```

```python
# Create a bar plot to visualize the cancellation rate
plt.figure(figsize=(10, 6))
plt.bar(['Not Canceled', 'Canceled'], [100 - cancellation_rate, cancellation_rate], color=['green', 'red'])
plt.title('Booking Cancellation Rate')
plt.ylabel('Percentage')
plt.show()
```
