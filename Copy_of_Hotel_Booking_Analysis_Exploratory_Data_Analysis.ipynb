# **Questions**
1) What is the overall distribution of bookings?
---



```python
# DataFrame is named data
# Convert 'arrival_date_month' to a categorical variable with a custom order
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
data['arrival_date_month'] = pd.Categorical(data['arrival_date_month'], categories=month_order, ordered=True)

# Calculate daily bookings
daily_bookings = data.groupby('arrival_date_day_of_month')['hotel'].count()

# Calculate monthly bookings
monthly_bookings = data.groupby('arrival_date_month')['hotel'].count()

# Plotting
plt.figure(figsize=(15, 6))

# Daily Bookings
plt.subplot(1, 2, 1)
daily_bookings.plot(kind='bar', color='skyblue')
plt.title('Daily Distribution of Bookings')
plt.xlabel('Day of Month')
plt.ylabel('Number of Bookings')

# Monthly Bookings
plt.subplot(1, 2, 2)
monthly_bookings.plot(kind='bar', color='salmon')
plt.title('Monthly Distribution of Bookings')
plt.xlabel('Month')
plt.ylabel('Number of Bookings')

plt.tight_layout()
plt.show()
```

# **2) What is the cancellation rate?**

```python
# The column 'is_canceled' contains binary values (0 for not canceled, 1 for canceled)
cancellation_rate = data['is_canceled'].mean() * 100
print(f"Cancellation Rate: {cancellation_rate:.2f}%")   #two floating point number

```

```python
# Create a bar plot to visualize the cancellation rate
plt.figure(figsize=(10, 6))
plt.bar(['Not Canceled', 'Canceled'], [100 - cancellation_rate, cancellation_rate], color=['green', 'red'])
plt.title('Booking Cancellation Rate')
plt.ylabel('Percentage')
plt.show()
```

# **3) What types of rooms are booked most frequently?**

```python
room_type_frequency = data['reserved_room_type'].value_counts()
room_type_frequency
```

```python
# Set the plotting style
sns.set(style="darkgrid")

# Create a count plot for room types
plt.figure(figsize=(12, 6))
sns.countplot(x='reserved_room_type', data=data, palette='viridis')
plt.title('Distribution of Booked Room Types')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.show()
```

```python
# Room type ADR
room_adr = data.groupby('assigned_room_type')['adr'].mean().sort_values(ascending=False)

# Plotting the room ADR
plt.figure(figsize=(10, 6))
sns.barplot(x=room_adr.index, y=room_adr.values, palette='plasma')
plt.title('Average Daily Rate (ADR) by Room Type')
plt.xlabel('Room Type')
plt.ylabel('ADR')
plt.show()
```

# **4) What is percentage of bookings in each hotel?**

```python
# Replace 'data' with the actual variable name of your DataFrame

# Calculate the total number of bookings in each hotel
hotel_counts = data['hotel'].value_counts()

# Calculate the percentage of bookings in each hotel
percentage_bookings = (hotel_counts / len(data)) * 100
```

# **5)  How does the booking demand vary across different seasons?**

```python
# Assuming your DataFrame is named 'data'
# Create a new column for season based on the month (Indian seasons)
data['season'] = data['arrival_date_month'].map({
    'January': 'Winter',
    'February': 'Winter',
    'March': 'Summer',
    'April': 'Summer',
    'May': 'Summer',
    'June': 'Monsoon',
    'July': 'Monsoon',
    'August': 'Monsoon',
    'September': 'Autumn',
    'October': 'Autumn',
    'November': 'Pre-Winter',
    'December': 'Winter'
})

# Group by season and calculate the total number of bookings
seasonal_booking_counts = data.groupby('season')['reservation_status'].value_counts().unstack()

# Plot the data
sns.set(style="whitegrid")
sns.countplot(x='season', hue='reservation_status', data=data)



plt.title('Booking Demand Across Indian Seasons')
plt.xlabel('Season')
plt.ylabel('Number of Bookings')
plt.legend(title='Reservation Status', loc='upper right')
plt.show()
```

# **7) what is Average Daily Price month wise?**

```python
#DataFrame is named 'data'

# Filter data for each hotel type
city_hotel_data = data[data['hotel'] == 'City Hotel']
resort_hotel_data = data[data['hotel'] == 'Resort Hotel']

# Group data by 'arrival_date_month' and calculate the mean ADR for each month
city_avg_adr = city_hotel_data.groupby('arrival_date_month')['adr'].mean()
resort_avg_adr = resort_hotel_data.groupby('arrival_date_month')['adr'].mean()

# Set the order of months for proper sorting
month_order = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]

# Create a line plot
plt.figure(figsize=(12, 6))
sns.lineplot(x=city_avg_adr.index, y=city_avg_adr.values, label='City Hotel', sort=False)
sns.lineplot(x=resort_avg_adr.index, y=resort_avg_adr.values, label='Resort Hotel', sort=False)

# Set plot labels and title
plt.xlabel('Month')
plt.ylabel('Average Daily Rate (ADR)')
plt.title('Average Daily Rate (ADR) Month-wise for City and Resort Hotels')
plt.xticks(month_order, rotation=45)  # Set x-axis ticks to show month names

# Show legend
plt.legend()

# Show the plot
plt.show()
```
